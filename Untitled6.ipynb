```python
# Install IBM Fairness 360 and matplotlib
!pip install aif360 matplotlib

# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from aif360.datasets import BinaryLabelDataset
from aif360.metrics import ClassificationMetric
from aif360.algorithms.preprocessing import Reweighing

# Load dataset with error handling
try:
    # To upload 'loan_data.csv', you can uncomment and run the following lines:
    from google.colab import files
    uploaded = files.upload()
    if 'loan_data.csv' not in uploaded:
        raise FileNotFoundError("'loan_data.csv' not uploaded.")

    df = pd.read_csv('loan_data.csv')  # Replace with your actual path if not in working directory
    print("CSV file loaded successfully.")
except FileNotFoundError:
    print("Error: 'loan_data.csv' file not found. Please upload the file in the working directory.")
    raise
except Exception as e:
    print(f"Error loading CSV file: {e}")
    raise

df.dropna(inplace=True)

# Encode categorical variables
le_gender = LabelEncoder()
le_race = LabelEncoder()
le_employment = LabelEncoder()

df['gender'] = le_gender.fit_transform(df['gender'])  # 0 = Female, 1 = Male
df['race'] = le_race.fit_transform(df['race'])      # e.g., 0 = Black, 1 = White, etc.
df['employment_status'] = le_employment.fit_transform(df['employment_status'])

# Define features and label
features = ['income', 'credit_score', 'employment_status']
X = df[features]
y = df['loan_approved']
protected_gender = df['gender']
protected_race = df['race']
protected_employment = df['employment_status'] # New: Capture employment_status as protected attribute

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create BinaryLabelDataset
dataset = BinaryLabelDataset(
    favorable_label=1,
    unfavorable_label=0,
    df=pd.DataFrame(X_scaled, columns=features).assign(
        loan_approved=y,
        gender=protected_gender,
        race=protected_race,
        employment_status=protected_employment # New: Add employment_status to the DataFrame
    ),
    label_names=['loan_approved'],
    protected_attribute_names=['gender', 'race', 'employment_status'] # New: Include employment_status here
)

# Split into train/test
train, test = dataset.split([0.7], shuffle=True)

# Train baseline model
model = LogisticRegression()
model.fit(train.features, train.labels.ravel())
y_pred = model.predict(test.features)

# Create predicted dataset
predicted_dataset = test.copy()
predicted_dataset.labels = y_pred

# Fairness metrics (gender)
gender_metric = ClassificationMetric(test, predicted_dataset,
                                     unprivileged_groups=[{'gender': 0}],
                                     privileged_groups=[{'gender': 1}])

# Fairness metrics (race)
race_metric = ClassificationMetric(test, predicted_dataset,
                                   unprivileged_groups=[{'race': 0}],
                                   privileged_groups=[{'race': 1}])

# Print baseline results
print("Gender Bias Audit:")
print("Accuracy:", accuracy_score(test.labels, y_pred))
print("Disparate Impact:", gender_metric.disparate_impact())
print("Equal Opportunity Difference:", gender_metric.equal_opportunity_difference())
print("Average Odds Difference:", gender_metric.average_odds_difference())
print("Statistical Parity Difference:", gender_metric.statistical_parity_difference())

print("Race Bias Audit:")
print("Disparate Impact:", race_metric.disparate_impact())
print("Equal Opportunity Difference:", race_metric.equal_opportunity_difference())
print("Average Odds Difference:", race_metric.average_odds_difference())
print("Statistical Parity Difference:", race_metric.statistical_parity_difference())

# Mitigation: Reweighing
rw = Reweighing(unprivileged_groups=[{'gender': 0}], privileged_groups=[{'gender': 1}])
train_transformed = rw.fit_transform(train)

# Retrain model on mitigated data
model_rw = LogisticRegression()
model_rw.fit(train_transformed.features, train_transformed.labels.ravel())
y_rw_pred = model_rw.predict(test.features)

# Evaluate mitigated predictions
predicted_rw_dataset = test.copy()
predicted_rw_dataset.labels = y_rw_pred

# Fairness metrics after mitigation
gender_metric_rw = ClassificationMetric(test, predicted_rw_dataset,
                                        unprivileged_groups=[{'gender': 0}],
                                        privileged_groups=[{'gender': 1}])

# Print mitigated results
print("After Mitigation (Reweighing):")
print("Accuracy:", accuracy_score(test.labels, y_rw_pred))
print("Disparate Impact:", gender_metric_rw.disparate_impact())
print("Equal Opportunity Difference:", gender_metric_rw.equal_opportunity_difference())
print("Average Odds Difference:", gender_metric_rw.average_odds_difference())
print("Statistical Parity Difference:", gender_metric_rw.statistical_parity_difference())

# Visualization of fairness metrics before and after mitigation
metrics_names = [
    'Disparate Impact',
    'Equal Opportunity Difference',
    'Average Odds Difference',
    'Statistical Parity Difference'
]

baseline_metrics = [
    gender_metric.disparate_impact(),
    gender_metric.equal_opportunity_difference(),
    gender_metric.average_odds_difference(),
    gender_metric.statistical_parity_difference()
]

mitigated_metrics = [
    gender_metric_rw.disparate_impact(),
    gender_metric_rw.equal_opportunity_difference(),
    gender_metric_rw.average_odds_difference(),
    gender_metric_rw.statistical_parity_difference()
]

x = np.arange(len(metrics_names))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, baseline_metrics, width, label='Before Mitigation')
rects2 = ax.bar(x + width/2, mitigated_metrics, width, label='After Mitigation')

ax.set_ylabel('Metric Value')
ax.set_title('Fairness Metrics Before and After Mitigation (Gender)')
ax.set_xticks(x)
ax.set_xticklabels(metrics_names, rotation=15, ha='right')
ax.legend()

for rect in rects1 + rects2:
    height = rect.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Mitigation: Reweighing for Employment Status
rw_employment = Reweighing(unprivileged_groups=unprivileged_employment, privileged_groups=privileged_employment)
train_transformed_employment = rw_employment.fit_transform(train)

# Retrain model on employment-status-mitigated data
model_rw_employment = LogisticRegression()
model_rw_employment.fit(train_transformed_employment.features, train_transformed_employment.labels.ravel())
y_rw_pred_employment = model_rw_employment.predict(test.features)

# Evaluate employment-status-mitigated predictions
predicted_rw_dataset_employment = test.copy()
predicted_rw_dataset_employment.labels = y_rw_pred_employment

# Fairness metrics after employment status mitigation
employment_metric_rw = ClassificationMetric(test, predicted_rw_dataset_employment,
                                          unprivileged_groups=unprivileged_employment,
                                          privileged_groups=privileged_employment)

# Print mitigated results for Employment Status
print("\nAfter Mitigation (Reweighing) for Employment Status:")
print("Accuracy:", accuracy_score(test.labels, y_rw_pred_employment))
print("Disparate Impact:", employment_metric_rw.disparate_impact())
print("Equal Opportunity Difference:", employment_metric_rw.equal_opportunity_difference())
print("Average Odds Difference:", employment_metric_rw.average_odds_difference())
print("Statistical Parity Difference:", employment_metric_rw.statistical_parity_difference())
```

### Loan Approval Distribution by Key Attributes

These visualizations show the count of loan approvals and rejections across different categories for protected attributes (Gender, Race, Employment Status) and key financial features (Income, Credit Score). This helps in understanding the underlying distribution of the 'loan_approved' label within different segments of the population.

```python
import seaborn as sns

# Gender vs. Loan Approved
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x=df['gender'].map(dict(enumerate(le_gender.classes_))), hue='loan_approved', palette='viridis')
plt.title('Loan Approvals by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Applicants')
plt.legend(title='Loan Approved', labels=['Rejected', 'Approved'])
plt.tight_layout()
plt.show()

# Race vs. Loan Approved
plt.figure(figsize=(9, 6))
sns.countplot(data=df, x=df['race'].map(dict(enumerate(le_race.classes_))), hue='loan_approved', palette='viridis')
plt.title('Loan Approvals by Race')
plt.xlabel('Race')
plt.ylabel('Number of Applicants')
plt.legend(title='Loan Approved', labels=['Rejected', 'Approved'])
plt.tight_layout()
plt.show()

# Employment Status vs. Loan Approved
plt.figure(figsize=(9, 6))
sns.countplot(data=df, x=df['employment_status'].map(dict(enumerate(le_employment.classes_))), hue='loan_approved', palette='viridis')
plt.title('Loan Approvals by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Number of Applicants')
plt.legend(title='Loan Approved', labels=['Rejected', 'Approved'])
plt.tight_layout()
plt.show()

# Income vs. Loan Approved (binned)
df['income_bin'] = pd.cut(df['income'], bins=5, labels=False)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='income_bin', hue='loan_approved', palette='viridis')
plt.title('Loan Approvals by Income Bin')
plt.xlabel('Income Bin')
plt.ylabel('Number of Applicants')
plt.legend(title='Loan Approved', labels=['Rejected', 'Approved'])
plt.tight_layout()
plt.show()

# Credit Score vs. Loan Approved (binned)
df['credit_score_bin'] = pd.cut(df['credit_score'], bins=5, labels=False)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='credit_score_bin', hue='loan_approved', palette='viridis')
plt.title('Loan Approvals by Credit Score Bin')
plt.xlabel('Credit Score Bin')
plt.ylabel('Number of Applicants')
plt.legend(title='Loan Approved', labels=['Rejected', 'Approved'])
plt.tight_layout()
plt.show()
```
